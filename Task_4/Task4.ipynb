# Task 4: Superstore Data Analysis

In Task 4, I performed a complete exploratory analysis on the Superstore dataset, focusing on sales, profit, orders, and their distribution across categories, sub-categories, regions, and shipping modes.

### Step 1: Import and Load Data
I imported the dataset using Pandas and loaded the CSV file. I also checked the structure and information of the data to understand columns, data types, and sample records.

### Step 2: Quick Look at the Data
I viewed the first few rows and examined the overall structure of the dataset to identify key columns such as Sales, Profit, Category, Sub-Category, Region, and Ship Mode.

### Step 3a: Total Sales by Region
I calculated the total sales for each region to identify which regions generate the highest revenue.

### Step 3b: Average Profit by Category
I computed the average profit for each product category to understand which categories are the most profitable.

### Step 3c: Orders by Ship Mode
I counted the number of orders for each shipping mode to see which shipping options are most frequently used by customers.

### Step 3d: Sales and Profit by Sub-Category
I aggregated sales and profit at the sub-category level to identify which specific product types contribute most to revenue and profit.

### Step 3e: Multi-level Grouping (Region + Category)
I performed a multi-level grouping of sales by both region and category to analyze patterns in different regions for each product category.

### Visual Analysis
For each step, I created enhanced bar charts using Seaborn and Matplotlib to visually represent the results, making it easier to identify trends and insights in the data.

```python
import pandas as pd
```

```python
df = pd.read_csv("superstore.csv", encoding="latin1")
```

```python
print(df.columns.tolist())
```

```python
sales_by_region = df.groupby("Region")["Sales"].sum().reset_index()
print("Total Sales by Region:\n", sales_by_region)
```

```python
avg_profit_by_category = df.groupby("Category")["Profit"].mean().reset_index()
print("Average Profit by Category:\n", avg_profit_by_category)
```

```python
orders_by_shipmode = df.groupby("Ship Mode")["Order ID"].count().reset_index()
orders_by_shipmode.rename(columns={"Order ID": "Order Count"}, inplace=True)
print("Orders by Ship Mode:\n", orders_by_shipmode)
```

```python
sales_profit_subcat = df.groupby("Sub-Category")[["Sales", "Profit"]].sum().reset_index()
print("Sales and Profit by Sub-Category:\n", sales_profit_subcat)
```

```python
region_category_sales = df.groupby(["Region", "Category"])["Sales"].sum().reset_index()
print("\nSales by Region and Category:\n", region_category_sales)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(data=sales_by_region.sort_values("Sales", ascending=False),
            x="Region", y="Sales", hue="Region", palette="Blues_d", legend=False)
plt.title("Total Sales by Region", fontsize=14, weight="bold")
plt.ylabel("Total Sales")
plt.xlabel("Region")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=avg_profit_by_category.sort_values("Profit", ascending=False),
            x="Category", y="Profit", hue="Category", palette="Greens_d", legend=False)
plt.title("Average Profit by Category", fontsize=14, weight="bold")
plt.ylabel("Average Profit")
plt.xlabel("Category")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=orders_by_shipmode.sort_values("Order Count", ascending=False),
            x="Ship Mode", y="Order Count", hue="Ship Mode", palette="Purples_d", legend=False)
plt.title("Order Count by Ship Mode", fontsize=14, weight="bold")
plt.ylabel("Order Count")
plt.xlabel("Ship Mode")
plt.show()
```

```python
sales_profit_melted = sales_profit_subcat.melt(id_vars="Sub-Category",
                                              value_vars=["Sales", "Profit"],
                                              var_name="Metric",
                                              value_name="Value")

plt.figure(figsize=(12,6))
sns.barplot(data=sales_profit_melted.sort_values("Value", ascending=False),
            x="Sub-Category", y="Value", hue="Metric", palette="Set2")
plt.title("Sales and Profit by Sub-Category", fontsize=14, weight="bold")
plt.ylabel("Value")
plt.xlabel("Sub-Category")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Metric")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=region_category_sales,
            x="Region", y="Sales", hue="Category", palette="Set3")
plt.title("Sales by Region and Category", fontsize=14, weight="bold")
plt.ylabel("Total Sales")
plt.xlabel("Region")
plt.legend(title="Category")
plt.show()
```
